```python
import glob
import time
from collections import defaultdict

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#"https://www.maxpreps.com/leaders/football/offense,passing/stat-leaders.htm"
#"https://www.maxpreps.com/leaders/football/offense,passing/stat-leaders-2.htm"

cats = [
    'offense,passing',
    'offense,rushing',
    'offense,receiving',
    'offense,scoring',
    'offense,touchdowns',
    'offense,total+yards',
    'defense,tackles',
    'defense,sacks',
    'defense,interceptions'
]



routes = [f"https://www.maxpreps.com/leaders/football{year_cond}/{cat}/stat-leaders{ending}.htm"
 for year_cond in ['', '-fall-19', '-fall-18']
 for cat in cats
 for ending in ['' if i==1 else f'-{i}' for i in range(1,11)]
]
```

```python
table_class_name = 'stat-leaders'
```

```python
sitemaps = {}
for i in routes:
    try:
        file_path = i.replace('https://www.maxpreps.com/leaders/', '').replace(',', '_').replace('/', '_')
        resp = requests.get(i)
        if resp.status_code != 200:
            print(f'Failed to query: {i}\nwith status: {resp.status_code}')
            continue
        fpath = f'./data/{file_path}'
        
        with open(fpath, 'w+') as fout:
            fout.write(resp.text)

        print(f'Wrote file to {fpath} for link {i}')
        sitemaps[i] = fpath
        time.sleep(5)
    except Exception as error:
        print(error, i)
```

```python
def translate_fpath(fpath):
    fname = fpath.split('/')[-1]
#     print(fname)
    if fname.startswith('football_'):
        file_year = 2020
    else:
        file_year = 2000 + int(fname.split('football-fall-')[1][:2])
    
    category = ','.join(fname.split('_')[1:3])
    return {
        'file_year': file_year,
        'category': category
    }
    
```

```python
dfs = defaultdict(lambda: defaultdict(list))
agg_dfs = defaultdict(lambda: defaultdict(list))
for fpath in glob.glob('./data/*.htm'):
#     print(fpath)
    with open(fpath, 'r') as fin:
        html_doc = fin.read()
        soup = BeautifulSoup(html_doc, 'html.parser')
        table = soup.findAll('table', {'class': table_class_name})
        data = [process_row(row) for row in table[0].find('tbody').find_all('tr')]
        path_dict = translate_fpath(fpath)
        dfs[path_dict['file_year']][path_dict['category']].append(pd.DataFrame([{**path_dict, **i} for i in data if i is not None]))

for year, year_dict in dfs.items():
    for cat, cat_list in year_dict.items():
        agg_dfs[year][cat] = pd.concat(cat_list)
```

```python
dfs_again = defaultdict(list)
for year, year_dict in agg_dfs.items():
    for cat, df in year_dict.items():
        dfs_again[cat].append(df)
```

```python
for cat, dfs in dfs_again.items():
    pd.concat(dfs).to_csv(f'players_{cat}.csv', index=False)
```

```python
def process_row(row):
    row_dict = dict()
    try:
        row_dict['rank'] = row.find_all('td', {'class': 'rank'})[0].get_text()
    except Exception as ex:
#         print(ex, 'rank')
        return None
    try:
        row_dict['name'] = row.find_all('th')[0].find('a').get_text()
    except Exception as ex:
#         print(ex, 'name')
        return None
    try:
        row_dict['year'] = row.find('abbr')['title']
    except Exception as ex:
#         print(ex, 'year', row)
        return None
    try:
        row_dict['high_school'] = row.find_all('a', {'class': 'team'})[0].get_text()  # a class="team
    except Exception as ex:
#         print(ex, 'high_school')
        return None
    try:
        row_dict['city_state'] = row.find_all('span', {'class': 'city-state'})[0].get_text()  # span class="city-state
    except Exception as ex:
#         print(ex, 'city_state')
        return None
    try:
        row_dict['position'] = row.find_all('td', {'class': 'position'})[0].get_text()
    except Exception as ex:
#         print(ex, 'position')
        return None
    
    try:
        classes = [td['class'][0] for td in row.find_all('td')]
        for _class in classes:
            row_dict[_class] = row.find_all('td', {'class': _class})[0].get_text()
    except Exception as ex:
#         print(ex, 'classes')
        return None

    return row_dict
```
